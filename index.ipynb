```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('C:/Users/Mayank/Desktop/project-data-sciencde/tracks.csv')
df_tracks.head()
```

```python
# check the total number of  null value in data set



pd.isnull(df_tracks).sum()
```

```python
#finding the total number of row and column & check the data type memory usage




df_tracks.info()
```

```python
# first major analysis 
# find the 5 least popular song present in the spotify  data set



sorted_df = df_tracks.sort_values('popularity', ascending = True).head(6)
sorted_df
```

```python
#let see the describe statistics for the numerical variable present in a column



df_tracks.describe().transpose()
```

```python
# 5 most popular song  which are greater then 70



most_popular= df_tracks.query('popularity>70' , inplace = False).sort_values('popularity', ascending = False)
most_popular[:5]
```

```python
# check the artist at the 50 th row in a data set 


df_tracks[["artists"]].iloc[10]
```

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1,center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# we are going to sample a data and take just quit 4 % of te total  data  and will create  to 2 represent plot using  this data 


sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#we are going to create a regression plot between  loudness and energy



plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy" , color = "c").set(title = "LOUDNESS S Energy correlation")
```

```python
#we are going to create a regression plot between  popularity and acousticness



plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness" , color = "b").set(title = "Popularity vs Acousticness correlation")
```
